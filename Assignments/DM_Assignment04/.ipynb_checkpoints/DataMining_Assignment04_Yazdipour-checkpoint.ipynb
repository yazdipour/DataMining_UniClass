```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split, KFold
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
import pydotplus 
from IPython.display import Image 
%matplotlib inline
```

# 1st Question

```python
nms=['id','Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape', 'Marginal Adhesion',   'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin', 'Normal Nucleoli', 'Mitoses','status']
ctest=pd.read_csv('CancerTest.txt',names=nms)
ctrain=pd.read_csv('CancerTrain.txt',names=nms)
```

```python
ctrain.head(2)
```

```python
ctest.head(2)
```

```python
X_train= ctrain.iloc[:,:10]
Y_train= ctrain.status
X_test= ctest.iloc[:,:10]
Y_test= ctest.status
```

```python
X_train.head(2)
```

```python
clf = DecisionTreeClassifier(criterion = "gini", max_depth=3, min_samples_leaf=4)
clf.fit(X_train, Y_train)
y_pred = clf.predict(X_test)
print("Accuracy of Decision Tree classifier with criterion + gini", accuracy_score(Y_test,y_pred)*100)
```

```python
clf2 = DecisionTreeClassifier(criterion = "entropy", max_depth=3, min_samples_leaf=3)
clf2.fit(X_train, Y_train)
y_pred2 = clf2.predict(X_test)
print("Accuracy of Decision Tree classifier with criterion + entropy ", accuracy_score(Y_test,y_pred2)*100)
```

# 2nd Question

```python
def readFile(fileName):
    f = open(fileName)
    lines=[];passedBlank=False;alls=[]
    for line in f:
        if passedBlank:
            if line == '\n':
                content = ' '.join(lines[3:])
                l=len(lines)
                if l==0:
                    alls.append(['','','','',''])
                elif l==1:
                    alls.append([lines[0],'','','',''])
                elif l==2:
                    alls.append([lines[0],lines[1],'','',''])
                elif l>2:
                    try:i = lines[2].rindex(',')
                    except:i=len(lines[2])
                    alls.append([lines[0],lines[1],lines[2][:i],lines[2][i:].replace(',','').replace('-',''),content])
                lines=[];content='';continue
        if line == '\n':
            passedBlank=True
        else:
            passedBlank=False
            lines.append(line.replace('\n',' '))
    f.close()
    return alls
```

```python
atrain=readFile('training.data')
atest=readFile('test.data')
```

```python
names=['topic', 'title', 'location','date','text']
df = pd.DataFrame(atrain, columns=names)
df.head(2)
```

```python
dft = pd.DataFrame(atest, columns=names)
dft.head(2)
```

```python
# from sklearn.feature_extraction.text import TfidfVectorizer
#     # Vectorise and TF-IDF transform the corpus 
#     X, y = create_tfidf_training_data(ref_docs)
```

```python
f=open('topics.data','r')
n_neighbors = 0
for line in f:
    n_neighbors += 1
f.close()
xtrain=df.iloc[df.index, 1:]
ytrain=df.iloc[df.index, 0]
xtest=dft.iloc[dft.index, 1:]
ytest=dft.iloc[dft.index, 0]
```

```python
def mydist(x, y):
    return 0 if (x in y) else 1
def kn():
#     clf = KNeighborsClassifier(n_neighbors, weights='distance')
    clf = KNeighborsClassifier(n_neighbors, algorithm='ball_tree', metric=mydist)
    clf.fit(xtrain, ytrain)
    ypred = clf.predict(xtest)
    print('Model Score with %d neigbors: %f' %(n_neighbors,accuracy_score(ytest,ypred)*100))
```
